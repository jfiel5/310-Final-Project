```python
# scrape_multi_players.py

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
import re

# --- Set up Chrome WebDriver ---
options = Options()
# options.add_argument("--headless")      # Uncomment to run in background
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# --- List of player-stats URLs ---
players = [
    "https://www.nba.com/stats/player/2544?SeasonType=Regular+Season&PerMode=Totals",
    "https://www.nba.com/stats/player/201939?PerMode=Totals",
    "https://www.nba.com/stats/player/977/career",
    "https://www.nba.com/stats/player/203999?SeasonType=Regular+Season&PerMode=Totals", 
    "https://www.nba.com/stats/player/406/career?PerMode=Totals",
    "https://www.nba.com/stats/player/201566?SeasonType=Regular+Season&PerMode=Totals",
    "https://www.nba.com/stats/player/203507?SeasonType=Regular+Season&PerMode=Totals"
]

# Will hold your DataFrames if you need them after the loop
all_dfs = {}
sheets = [] 

for url in players:
    driver.get(url)
    # give JS time to render the table
    time.sleep(6)

    # pull in the rendered DOM
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # find the first <table> on the page
    table = soup.find("table")
    if table is None:
        print(f"No table found at {url}")
        continue

    # pandas will parse out the HTML table for us
    df = pd.read_html(str(table))[0]

    # extract the player ID from the URL for naming
    name = driver.find_element(By.CLASS_NAME, "PlayerSummary_playerNameText___MhqC").text


    # save to CSV
    csv_name = f"player_{name}_stats.csv"
    df.to_csv(csv_name, index=False)
    print(f"✔️ Saved main table for player {name} to {csv_name}")
    sheets.append(csv_name)

    # stash the DataFrame if you need it later
    all_dfs[name] = df

driver.quit()

# optional: do something with all_dfs, e.g. display keys
print("Done. DataFrames created for:", list(all_dfs.keys()))

```

```python
from matplotlib import pyplot as plt
from matplotlib import style
import pandas as pd
import re

cleaned_sheets = []

for sheet in sheets:
    df = pd.read_csv(sheet, header=[0,1])
    def flatten(col):
        lvl0, lvl1 = col
        if "Unnamed" in lvl0:
            name = lvl1
        else:
            name = lvl1 if lvl1 and lvl1 not in [""] else lvl0
        return name.strip()
    df.columns = [flatten(col) for col in df.columns]
    df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
    df = df.rename(columns={
        "By Year": "Season",
        "TEAM":    "Team",
        })
    for col in df.columns.difference(["Season", "Team"]):
        df[col] = pd.to_numeric(df[col], errors="coerce")
    df = df.reset_index(drop=True)
    cleaned_sheets.append(df)
    print(df.dtypes)
    print(df.head())
```

```python
#visualize the data
import matplotlib.pyplot as plt
# Example: Plotting points per game for the first player
for i in range(len(cleaned_sheets)):
    player_df = cleaned_sheets[i]
    plt.figure(figsize=(10, 5))
    plt.plot(player_df['Season'], player_df['PTS'], marker='o', label='Points')
    plt.title(f"Points per Game Over Seasons for Player {i+1}")
    plt.xlabel('Season')
    plt.ylabel('Points per Game')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(8, 6))

for i, df in enumerate(cleaned_sheets, start=1):
    # Make sure FGA/FGM are numeric
    x = pd.to_numeric(df['FGA'], errors='coerce')  
    y = pd.to_numeric(df['FGM'], errors='coerce')
    
    plt.scatter(x, y, label=f'Player {i}')

plt.xlabel('Field Goals Attempted (FGA)')
plt.ylabel('Field Goals Made (FGM)')
plt.title('FGM vs FGA for Players')
plt.xlim(0, 2000)
plt.ylim(0, 1500)
plt.legend()
plt.tight_layout()
plt.show()
```
