```python
# scrape_multi_players.py

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
import re

options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# List of player-stats URLs
players = [
    "https://www.nba.com/stats/player/2544?SeasonType=Regular+Season&PerMode=Totals",
    "https://www.nba.com/stats/player/201939?PerMode=Totals",
    "https://www.nba.com/stats/player/977/career?PerMode=Totals",
    "https://www.nba.com/stats/player/203999?SeasonType=Regular+Season&PerMode=Totals", 
    "https://www.nba.com/stats/player/406/career?PerMode=Totals",
    "https://www.nba.com/stats/player/201566?SeasonType=Regular+Season&PerMode=Totals",
    "https://www.nba.com/stats/player/203507?SeasonType=Regular+Season&PerMode=Totals"
]

all_dfs = {}
sheets = [] 
names = []

for url in players:
    driver.get(url)
    time.sleep(5)

    # pull in the rendered DOM
    soup = BeautifulSoup(driver.page_source, "html.parser")

    table = soup.find("table")
    if table is None:
        print(f"No table found at {url}")
        continue

    df = pd.read_html(str(table))[0]

    # extract the player from the URL for naming
    name = str(driver.find_element(By.CLASS_NAME, "PlayerSummary_playerNameText___MhqC").text)


    names.append(name.replace("\n", " "))

    # save to CSV
    csv_name = f"player_{name}_stats.csv"
    df.to_csv(csv_name, index=False)
    print(f"Saved main table for player {name} to {csv_name}")
    sheets.append(csv_name)

    all_dfs[name] = df

driver.quit()

print("Done. DataFrames created for:", list(all_dfs.keys()))

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# cleaned_sheets = [df1, df2, df3, df4, df5, df6, …]
# Optional: if you have player names...
# names = ['Giannis','Curry','Player3','Player4','Player5','Player6']

# 1) Compute career total PTS for the first 6 players
totals = [
    pd.to_numeric(df['FG%'], errors='coerce').sum()
    for df in cleaned_sheets[:6]
]

# 2) Prepare labels
if 'names' in globals() and len(names) >= 6:
    labels = names[:6]
else:
    labels = [f'Player {i}' for i in range(1, 7)]

# 3) Plot bar chart
plt.figure(figsize=(10,5))
bars = plt.bar(labels, totals)

# 4) Annotate bars with the actual totals
for bar, val in zip(bars, totals):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + max(totals)*0.01,
             f'{int(val):,}', ha='center', va='bottom')

plt.title("Field Goal % by Player")
plt.ylabel("Total PTS")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# cleaned_sheets = [df1, df2, …]
# names = ['Giannis','Curry',…]  # or default Player 1…

# 1) Pull the raw FG% and 3P% (no *100!)
fg_vals = [pd.to_numeric(df['FG%'], errors='coerce').iloc[0] for df in cleaned_sheets[:6]]
tp_vals = [pd.to_numeric(df['3P%'], errors='coerce').iloc[0] for df in cleaned_sheets[:6]]

players = names[:6] if 'names' in globals() else [f'Player {i}' for i in range(1,7)]
df_pct = pd.DataFrame({'FG%': fg_vals, '3P%': tp_vals}, index=players)

# 2) Plot
ax = df_pct.plot(kind='bar', figsize=(10,6))
ax.set_ylabel('Percentage (%)')
ax.set_title('Latest Season FG% and 3P% by Player')
plt.xticks(rotation=45)

# 3) Annotate
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='edge')

plt.tight_layout()
plt.show()


```
